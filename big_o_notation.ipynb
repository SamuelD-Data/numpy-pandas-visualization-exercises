```python
# importing matplotlib with alias for future use
import matplotlib.pyplot as plt
 # importing for log values
import math
```

```python
# set x for all lines except for n^n
x = list(range(1, 4)) 
# separate x range set for n^n plot 
# because it reaches very much higher y values than the other lines and causes them to cluster as a result
# this is alleviated it by giving it a lower max x value so its y value doesnt get as large
x2 = list(range(1, 3))  

# set y values per specified formula
y1 = [n * 0 + 1 for n in x]
# plot both x and y, set line color and label for legend
plt.plot(x, y1, c = 'red', label = '$O(1)$')

# set y values per specified formula
y2 = [math.log(n) for n in x]
# plot both x and y, set line color and label for legend
plt.plot(x, y2, c = 'blue', label = '$O(log n)$') 

# set y values per specified formula
y3 = [n for n in x]
# plot both x and y, set line color and label for legend
plt.plot(x, y3, c = 'green', label = '$O(n)$')

# set y values per specified formula
y4 = [n * math.log(n) for n in x]
# plot both x and y, set line color and label for legend
plt.plot(x, y4, c = 'purple', label = '$O(n log n)$') 

# set y values per specified formula
y5 = [n ** 2 for n in x]
# plot both x and y, set line color and label for legend
plt.plot(x, y5, c = 'orange', label = '$O(n^2)$')

# set y values per specified formula
y6 = [2 ** n for n in x]
# plot both x and y, set line color and label for legend
plt.plot(x, y6, c = 'grey', label = '$O(2^n)$')

# set y values per specified formula
y7 = [n * 0 + 1 for n in x]
# plot both x and y, set line color and label for legend
plt.plot(x, y7, c = 'brown', label = '$O(n!)$')

# set y values per specified formula
y8 = [n ** n for n in x2]
# plot both x and y, set line color and label for legend
plt.plot(x2, y8, c = 'magenta', label = '$O(n^n)$')

# show legend
plt.legend(loc='upper left')

# add title
plt.title('$Big O Notation$')

# add x axis label
plt.xlabel('$Elements$')

# add y axis label
plt.ylabel('$Operations$')

# display
plt.show
```
