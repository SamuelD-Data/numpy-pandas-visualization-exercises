```python
import matplotlib.pyplot as plt # importing matplotlib with alias for future use
```

```python
"""
Use matplotlib to plot the following equation:
y = x ** 2 - x + 2
You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.
"""

# make x points via list and range 
x = list(range(-49, 50)) 
# make y points via list iteration and formula for x
y = [n ** 2 - n + 2 for n in x] 

# plot points
plt.plot(x,y)   
# add annotation details (location of text, color of arrow, etc.)
plt.annotate('Origin', xy=(0, 0), xytext=(-5, 550), 
             arrowprops={'facecolor': 'blue'})

# add title
plt.title('$y = x^2 - x + 2$')

# display 
plt.show()
```

```python
# create a plot for y = square root of x

# import math
import math

# make x points via list and range 
x = list(range(10, 100))
# make y points via list iteration and square root from math module
y = [math.sqrt(n) for n in x]

# plot points
plt.plot(x,y)   

# add title
plt.title('y = square root of x')

# display 
plt.show()
```

```python
# create a plot for y = x^3

# set x points
x = list(range(-10,10))

# set y points
y = [n ** 3 for n in x]

# plot points
plt.plot(x,y)

# add title
plt.title('y = x^3')

# display
plt.show()
```

```python
# create a plot for y = 2^x

# set x points
x = list(range(-100,100))

# set y points
y = [2 ** n for n in x]

# plot points
plt.plot(x,y)

# add title
plt.title('y = 2^x')

# display graph
plt.show()
```

```python
# combine the previous figures into one large figure with 4 subplots

# set number of rows and columns
n_rows = 2
n_cols = 2

# FIRST GRAPH

# set subplot position
plt.subplot(n_rows, n_cols, 1)
# set x points
x = list(range(-100,100))
# set y points
y = [2 ** n for n in x]
# plot points
plt.plot(x,y)
# add title
plt.title('y = $2^x$')

# SECOND GRAPH

# set subplot position 
plt.subplot(n_rows, n_cols, 2)
# set x points
x2 = list(range(-10,10))
# set y points
y2 = [n ** 3 for n in x2]
# plot points
plt.plot(x2,y2)
# add title
plt.title('y = $x^3$')

# THIRD GRAPH

# set subplot position 
plt.subplot(n_rows, n_cols, 3)
# make x points via list and range 
x3 = list(range(10, 100))
# make y points via list iteration and square root from math module
y3 = [math.sqrt(n) for n in x3]
# plot points
plt.plot(x3,y3)   
# add title
plt.title('y = square root of x')

# FOURTH GRAPH

# set subplot position 
plt.subplot(n_rows, n_cols, 4)
# make x points via list and range 
x4 = list(range(-49, 50)) 
# make y points via list iteration and formula for x
y4 = [n ** 2 - n + 2 for n in x4] 
# plot points
plt.plot(x4,y4)   
# add annotation details (location of text, color of arrow, etc.)
plt.annotate('Origin', xy=(0, 0), xytext=(-5, 550), 
             arrowprops={'facecolor': 'blue'})
# add title
plt.title('$y = x^2 - x + 2$')

# adjust spacing between graphs
plt.tight_layout()

# display
plt.show()
```

```python
# combine the 4 figures from the previous exercise into one large figure

# FIRST LINE

# set x points
x = list(range(-10,10))
# set y points
y = [2 ** n for n in x]
# plot points, specify line color, label line
plt.plot(x,y, c = 'blue', label='$2^x$')


# SECOND LINE

# set x points
x2 = list(range(-7,7))
# set y points
y2 = [n ** 3 for n in x2]
# plot points, specify line color, label line
plt.plot(x2,y2, c = 'red', label = '$x^3$')

# THIRD LINE

# make x points via list and range 
x3 = list(range(0, 10))
# make y points via list iteration and square root from math module
y3 = [math.sqrt(n) for n in x3]
# plot points, specify line color, label line
plt.plot(x3, y3, c = 'green', label = 'square root of x')   

# FOURTH LINE 

# make x points via list and range 
x4 = list(range(-20, 20)) 
# make y points via list iteration and formula for x
y4 = [n ** 2 - n + 2 for n in x4] 
# plot points, specify line color, label line
plt.plot(x4, y4, c = 'purple', label = '$y = x^2 - x + 2$')   

# adjust spacing between graphs
plt.tight_layout()

# show legend
plt.legend(loc='lower right')

# display
plt.show()
```
