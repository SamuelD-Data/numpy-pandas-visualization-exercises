```python
# Use the iris database to answer the following quesitons:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')

iris.to_csv('iris.csv')

iris = pd.read_csv('iris.csv')

iris = iris.drop(columns=['Unnamed: 0'])
```

```python
# What does the distribution of petal lengths look like?

sns.distplot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width?

sns.relplot(x = 'petal_width', y = 'petal_length', data = iris)

# Answer: we see that there is a positive relationship between petal length and petal width. on average, petals with shorter lengths have less width. 
# likewise, longer petals tend to be wider
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length?

sns.relplot(x = 'sepal_width', y = 'sepal_length', hue = 'species', data = iris)

# although not 100% accurate, we could still make reasonable predictions of species based on sepal width and length
# for example, iris' with sepal lengths under 4.7 and sepal widths above 2.6 are probably setosas. 
```

```python
# Which features would be best used to predict species?

plt.figure(figsize=(13, 10))
plt.suptitle('Iris Species Identification')

plt.subplot(221)
sns.scatterplot(x = "petal_width", y = "petal_length", hue = 'species', data = iris)

plt.subplot(222)
sns.scatterplot(x = "sepal_width", y = "sepal_length", hue = 'species', data = iris)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Petal length and width seem to be better suited for use in predicting species. 
# This is due to the fact that the data points for petal L and W are more seperated between species than sepal attributes.
# For example, whereas the mid the high setosas lengths are within the range of many versicolor and some virginica data points
# we see that their petal data points hae a very large gap from any other species L or W. Although the versicolor and virginica
# species petal data points are closer than the setosa, they are still more seperated than their sepal counterparts.
```

```python
# Using the lesson as an example, use seaborn's load_dataset function to load the anscombe data set. 

anscombe = sns.load_dataset('anscombe')

anscombe.to_csv('anscombe.csv')

anscombe = pd.read_csv('anscombe.csv')

anscombe = anscombe.drop(columns=['Unnamed: 0'])

anscombe
```

```python
# Use pandas to group the data by the dataset column, and calculate summary statistics for each dataset. 
# What do you notice?

x_summary = anscombe.groupby('dataset').x.describe()

y_summary = anscombe.groupby('dataset').y.describe()

xy_summary =pd.merge(x_summary, y_summary, left_on='dataset', right_on='dataset', how='inner')

xy_summary

# Answer: For the most part, the summary statistics for each data set are either identical or very similar. For example, the count and mean of x from all 4 datasets is identical.
# The x data summary statistics for datasets 1-3 are completely identical. The y data for each data set is also very similar.
```
