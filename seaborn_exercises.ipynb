```python
# Importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy 
from pydataset import data
```

```python
# Load the iris dataset 

# loading iris data set and assigning to variable
iris = sns.load_dataset('iris')
```

```python
# What does the distribution of petal lengths look like?

# setting size for figure
plt.figure(figsize=(8, 4))

# creating distribution plot of petal lengths
sns.distplot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width?

# ploting scatterplot with x as petal width and y as petal length
sns.scatterplot(x = 'petal_width', y = 'petal_length', data = iris)

# Answer: we see that there is a positive relationship between petal length and petal width. on average, petals with shorter lengths have less width. 
# likewise, longer petals tend to be wider
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length?

# plotting scatterplot with x as sepal width and y as sepal length
# setting hue to species so that the data points for each species will be different colors
sns.scatterplot(x = 'sepal_width', y = 'sepal_length', hue = 'species', data = iris)

# although not 100% accurate, we could still make reasonable predictions of species based on sepal width and length
# for example, iris' with sepal lengths under 4.7 and sepal widths above 2.6 are probably setosas. 
```

```python
# Which features would be best used to predict species?

# setting figure size and title
plt.figure(figsize=(13, 10))
plt.suptitle('Iris Species Identification')

# setting subplot position 
plt.subplot(221)
# plotting scatterplot with x as petal width and y as petal length
# hue is set to species so that the data points of each species are different colors
sns.scatterplot(x = "petal_width", y = "petal_length", hue = 'species', data = iris)

# setting subplot position 
plt.subplot(222)
# plotting scatterplot with x as petal width and y as petal length
# hue is set to species so that the data points of each species are different colors
sns.scatterplot(x = "sepal_width", y = "sepal_length", hue = 'species', data = iris)

# adjusting position of plot legend so that it doesnt cover any data points
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Petal length and width seem to be better suited for use in predicting species. 
# This is due to the fact that the data points for petal L and W are more seperated between species than sepal attributes.
# For example, whereas the mid the high setosas lengths are within the range of many versicolor and some virginica data points
# we see that their petal data points hae a very large gap from any other species L or W. Although the versicolor and virginica
# species petal data points are closer than the setosa, they are still more seperated than their sepal counterparts.
```

```python
# Load the swiss dataset and read it's documentation.

# load swiss dataset
swiss = data('swiss')
# show swiss dataset documentation
data('swiss', show_doc = True)
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)

# Any area with that is equal to or more than 60 percent catholic will be considered a catholic area
# adding column to dataframe that shows whether an area is considered catholic
swiss['is_catholic'] = swiss.Catholic >= 60

swiss
```

```python
# Does whether or not a province is Catholic influence fertility?

# set plot size
plt.figure(figsize=(6, 5))

# create scatterplot
# x axis shows what percentage of the population is catholic while the y axis shows the fertility rate
sns.scatterplot(y = 'Fertility', x = 'Catholic', hue = 'is_catholic', data = swiss)

# set x and y labels along with plot title
plt.title('Fertility Rates: Catholic vs. Non-Catholic')
plt.xlabel('Catholic Percentage')
plt.ylabel('Fertility Rate')

# change position of legend to avoid covering data points
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Answer: The plot suggest that there may be a correlation between regions that are populated with more catholic than non-catholic residents
# All areas with fertility rates above 80% have a high percentage catholics within them. Furthermore, there are no catholic areas that have a fertility rate below 60%
```

```python
# What measure correlates most strongly with fertility?

# set figure size
plt.figure(figsize=(15, 15))

# set plot title
plt.suptitle('Fertility Rates')

# create lm plots to see how strong of a correlation different measures have with fertility
sns.lmplot(y = "Fertility", x = "Agriculture", data = swiss)
sns.lmplot(y = "Fertility", x = "Examination", data = swiss)
sns.lmplot(y = "Fertility", x = "Education", data = swiss)
sns.lmplot(y = "Fertility", x = "Catholic", data = swiss)
sns.lmplot(y = "Fertility", x = "Infant.Mortality", data = swiss)

# Answer: Based on the lmplots below, it appears that examination is most strongly tied to fertility with being catholic being a close second.
# although education also appears to have a strong correlation as well, there are too few data points for highly educated data points to trust it.
```

```python
# creating function with 4 parameters, 1 for each of the pieces of info we need to make a url with
def get_db_url(db):
    # importing host, username, and password from external file
    from env import host, username, password
    # setting variable equal to string that will combine the host, username, password, and db for our URL
    url = (f'mysql+pymysql://{username}:{password}@{host}/{db}')
    # return the URL that is created
    return (url)

# using get_db_url to explore to chipotle database
chipotle = pd.read_sql('SELECT * FROM orders', get_db_url('chipotle'))
```

```python
# Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and the revenue produced by each.

# creating dataframe that groups each item name, sums the quanity and only keeps the 4 largest
# in other words, the top 4 most ordered items
pop_items = chipotle.groupby('item_name').quantity.agg('sum').nlargest(4, keep = 'all')

# remove $ from cost column and convert to float so mathematical operations can be performed
chipotle_money = chipotle.assign(money_price = chipotle['item_price'].str.replace('$','').astype(float))

# create dataframe that groups items by name and sums the amount of revenue that each generated
rev_items = chipotle_money.groupby('item_name').money_price.agg('sum')

# merging dataframes together so we can see the amount of revenue that the top 4 most popular items generated
pop_and_rev = pd.merge(pop_items, rev_items, left_on = 'item_name', right_on = 'item_name', how = 'inner')

# setting figure size
plt.figure(figsize=(8, 4))

# setting plot title
plt.title('Revenue for Top 4 Items at Chipotle')

# creating barplot to display findings
sns.barplot(x = pop_and_rev.index, y = 'money_price',  data = pop_and_rev)

# setting x and y labels
plt.xlabel(' Item Name')
plt.ylabel('Revenue')
```

```python
# Load the sleepstudy data and read it's documentation. 

# loading dataset
sleep = data('sleepstudy')

# showing dataset documentation
data('sleepstudy', show_doc = True)
```

```python
# Use seaborn to create a line chart of all the individual subject's 
# reaction times and a more prominant line showing the average change in reaction time.

# create data frame with data grouped by days that shows the average reaction time per day
avg_sleep = pd.DataFrame(sleep.groupby('Days').Reaction.agg('mean'))

# set figure size
plt.figure(figsize=(12, 10))

# plot average for each day as line, set linewidth to 7 so it stands out, add label for legend, change color to purple
plt.plot(avg_sleep.index, avg_sleep.Reaction, linewidth = 5, linestyle = '--', label = 'average', c = 'purple')

# add average line to legend
plt.legend('average')

# plot each patients reaction times, color by subject number, use palette that gives each patient a unique color
sns.lineplot('Days', 'Reaction', hue = 'Subject', data = sleep, palette = 'Paired')

# change position of legend to avoid covering data points
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```
