```python
# Copy the code from the lesson to create a dataframe full of student grades.
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
# Create a column named passing_english that indicates whether each student has a passing grade in reading.

# creating column that will hold True for a student if their english grade is equal to or greater than 70 and False otherwise
df['passing_english'] = df.english >= 70

df
```

```python
# Sort the english grades by the passing_english column. 

#How are duplicates handled? - What?

# sorting by passing english column, False at top by default using sort values
df.sort_values(by = 'passing_english')
```

```python
# Sort the english grades first by passing_english and then by student name. All the students that are failing english should be first, and within the students that are failing english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)

# sorting by passing english first then name using sort values
df.sort_values(by = ['passing_english', 'name'])
```

```python
# Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.

# sorting by passing english first then english grade using sort values 
df.sort_values(by = ['passing_english', 'english'])
```

```python
# Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

# adding english, math and reading grades together then dividing by 3 to get average grade
df['total_grade'] = round((df.english + df.math + df.reading) / 3)

df
```

```python
# Rename the cty column to city.

# renaming cty column using rename
# using inplace to alter original dataframe 
mpg.rename(columns = {"cty" : "city"}, inplace = True)

mpg
```

```python
# Rename the hwy column to highway.

# renaming hwy column using rename
# using inplace to alter original dataframe 
mpg.rename(columns = {"hwy" : "highway"}, inplace = True)

mpg
```

```python
# Do any cars have better city mileage than highway mileage?

# using > to see if there are any city mileages that are greater than their respective highway mileages
mpg[mpg.city > mpg.highway]

# Answer: No because there are no cars with higher city mpg than highway mpg
```

```python
# Create a column named mileage_difference this column should contain the difference between highway and city mileage for each car.

# creating column with data being set to highway mileage minus city mileage
mpg['mileage_difference'] = mpg.highway - mpg.city

mpg
```

```python
# Which car (or cars) has the highest mileage difference?

# using max to get highest mileage difference
# filtering for cars with mileage difference that match the max
mpg[mpg.mileage_difference == mpg.mileage_difference.max()]
```

```python
# Which compact class car has the lowest highway mileage? 

# creating dataframe of all compact cars
compacts = mpg[mpg['class'] == "compact"]

# filtering new dataframe for cars with highway mileage that matches the minimum highway mileage in the dataframe found via min
lowest_hway_compacts = compacts[compacts['highway'] ==  compacts['highway'].min()]

lowest_hway_compacts
```

```python
# Which compact class car has the best highway mileage? 

# creating dataframe of all compact cars
compacts = mpg[mpg['class'] == "compact"]

# filtering new dataframe for cars with highway mileage that matches the maximum highway mileage in the dataframe found via max
best_hway_compacts = compacts[compacts['highway'] ==  compacts['highway'].max()]

best_hway_compacts
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.

# creating column that holds the average of the highway and city mileages
mpg['average_mileage'] = (mpg.highway + mpg.city) / 2

mpg
```

```python
# Which dodge car has the best average mileage? 

# creating dataframe that contains all dodge cars
dodges = mpg[mpg.manufacturer == 'dodge']

# filtering for cars with average mileage equal to the max average found in the dataframe via max
best_avg_miles_dodge = dodges[dodges.average_mileage == dodges.average_mileage.max()]

best_avg_miles_dodge
```

```python
# Which dodge car has the worst average mileage? 

# creating dataframe that contains all dodge cars
dodges = mpg[mpg.manufacturer == 'dodge']

# filtering for cars with average mileage equal to the mmin average found in the dataframe via min
worst_avg_miles_dodge = dodges[dodges.average_mileage == dodges.average_mileage.min()]

worst_avg_miles_dodge
```
